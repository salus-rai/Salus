```python
import collections
import collections.abc
from pptx import Presentation
from PIL import Image,ImageFilter
import io

def blur_header_footer(prs,output_ppt_path):
    #prs=Presentation(pptx_path)
    for slide in prs.slides:
        for shape in list(slide.shapes):
            if shape.shape_type==13:
                left,top,width,height=shape.left,shape.top,shape.width,shape.height
                slide_height=prs.slide_height
                if top<slide_height*0.2 or top+height>slide_height*0.8:
                    ima_path = Image.open(io.BytesIO(shape.image.blob))
                    blurred_image = ima_path.filter(ImageFilter.GaussianBlur(radius=15))
                    blurred_image_bytes = io.BytesIO()
                    blurred_image.save(blurred_image_bytes, format='PNG')
                    blurred_image_bytes.seek(0)
                    slide.shapes._spTree.remove(shape._element)
                    slide.shapes.add_picture(blurred_image_bytes, left, top, width, height)
                    #prs.save(output_ppt_path)

input_pptx_path=Presentation("C:/Users/shraddha.b03/IMPACT/MaskingPpt/MaskingPpt/MaskingPpt/inputFolder/sample2.pptx")
output_pptx_path="C:/Users/shraddha.b03/IMPACT/MaskingPpt/MaskingPpt/MaskingPpt/outputFolder/output5.pptx"
blur_header_footer(input_pptx_path,output_pptx_path)
input_pptx_path.save(output_pptx_path)
```
